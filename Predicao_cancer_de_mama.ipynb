### DataPrep

```python
data['Unnamed: 32'].fillna(data['Unnamed: 32'].mean(),inplace=True)
```

### Selecionar as vari√°veis

```python
data_filtro = data[["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean","radius_se","texture_se","perimeter_se","area_se","smoothness_se","compactness_se","concavity_se","concave points_se","symmetry_se","fractal_dimension_se","radius_worst","texture_worst","perimeter_worst","area_worst","smoothness_worst","compactness_worst","concavity_worst","concave points_worst","symmetry_worst","fractal_dimension_worst",]]
Var_resp = data["diagnosis"]
data = data_filtro
```

### Separar Treino e Teste

```python
from sklearn.model_selection import train_test_split
```

```python
X_all = data_filtro
y_all = Var_resp

X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size = 0.20 ,random_state = 84621)

print('Numero de observa√ßoes do treino:', len(X_train))
print('Numero de observa√ßoes da teste:',len(X_test))
```

# MODELAGEM E VALIDA√á√ÉO DOS RESULTADOS

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
clf = DecisionTreeClassifier(random_state=123,max_depth=3,min_samples_leaf=50)
```

```python
modelo = clf.fit(X_train,y_train)
```

### Valida√ß√£o cruzada com 5 folds e verificar a acur√°cia para cada fold

```python
scores = cross_val_score(clf, X_train, y_train, cv=5)
scores
```

## Verificando quais foram as colunas determinantes para o modelo

```python
importancia = clf.feature_importances_
importancia = pd.DataFrame(importancia)
importancia
```

```python
importancia = pd.DataFrame(importancia)

colunas = X_train.columns
colunas = pd.DataFrame(colunas)

importancia_concatenada = pd.concat([colunas,importancia],axis=1)
```

```python
importancia_concatenada
```

```python
importancia_concatenada.columns = ['coluna', 'importancia']
importancia_concatenada.sort_values(by=['importancia'], axis=0, ascending=False)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
